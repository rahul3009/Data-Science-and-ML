```python
import pandas as pd
```

```python
df = pd.read_csv('chips.tsv',sep='\t')
```

```python
df.item_price[0]
```

### How many products cost more than $10.00?

```python
df.item_price[0]
```

```python

def change(x):
    x=x.strip().replace('$','')
    
    return float(x)
```

```python
df.dtypes
df['item_price'] = df.item_price.apply(change)
```

```python
df[df['item_price']>10]
```

### print a data frame with only two columns item_name and item_price with no duplicates item and having 1 quantity

```python
# delete the duplicates in item_name
df_new = df.drop_duplicates(['item_name'])

# select only the products with quantity equals to 1
df_one_prod = df_new[df_new.quantity == 1]

df_n=df_one_prod.loc[:,['item_name','item_price']]

df_n.sort_values('item_price')
```
